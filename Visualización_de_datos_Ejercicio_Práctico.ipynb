```python
!pip install yfinance

import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import yfinance as yf

%matplotlib inline
```

```python
lista = ['MSFT', 'AAPL', 'AMZN']
```

```python
#Parametros de entrada para graficar
start = "2019-01-01"
end = '2022-1-01'
#Descargamos los datos de las cotizaciones de 4 empresas
msft = yf.download('MSFT',start,end)
aapl = yf.download('AAPL',start,end)
amzn = yf.download('AMZN',start,end)
```

```python
msft.info()
```

```python
msft.head()
```

```python
msft['Open'].plot(label = 'Microsoft', figsize = (15,7))
aapl['Open'].plot(label = "Apple")
amzn['Open'].plot(label = 'Amazon')
plt.title('Precio de las acciones de Microsoft, Apple y Amazon')
plt.legend()
```

```python
msft['Volume'].plot(label = 'Microsoft', figsize = (15,7))
aapl['Volume'].plot(label = "Apple")
amzn['Volume'].plot(label = 'Amazon')
plt.title('Volumen de acciones intercambiadas en el mercado')
plt.legend()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig2 = make_subplots(specs=[[{"secondary_y": True}]])
fig2.add_trace(go.Scatter(x=aapl.index,y=aapl['Close'],name='Precio'),secondary_y=False)
fig2.add_trace(go.Bar(x=aapl.index,y=aapl['Volume'],name='Volumen de operaciones'),secondary_y=True)
fig2.update_layout(title_text="Precio y Volumen de las operaciones de Apple en el mercado estadounidense")
fig2.show()
```
